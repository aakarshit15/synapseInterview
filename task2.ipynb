### The previous commands were just to make sure we have a workable dataframe now, we'll do some manipulations on the data.

Find out the top 5 most voted candidates in Telangana

```python
# Top 5 voted candidates in Telangana
# 1. Getting dataframe for Telangana
telanganaDF = df[df["State"] == "Telangana"]
# 2. Sorting data based on total votes in descending order
sortedTelanganaDF = telanganaDF.sort_values(by=["Total Votes"], ascending=False)
# 3. Getting top 5 voted candidates
sortedTelanganaDF.head(5)
```

Find out the total number of votes received by each state

```python
# Getting unique states
states = set(df["State"].tolist())

# Initalising voteCount[]
voteCount = []

# Calculating vote count for each state
for state in states: 
    totalVotes = 0
    candVotes = df[df["State"] == state]["Total Votes"] #<class 'pandas.core.series.Series'>
#     Adding each candidate's vote
    for candVote in candVotes:
        totalVotes += candVote
    # Appending data for a state
    voteCount.append({"State": state, "Votes": totalVotes})

# Converting voteCount to a data frame
voteCountDF = pd.DataFrame.from_dict(voteCount)

# Sorting voteCoungDF in descending order
voteCountDF.sort_values(by="Votes", ascending=False, inplace=True, ignore_index=True)

# Displaying voteCoutDF
voteCountDF
```

Show these values as a percentage of the total number of votes received in the election.

```python
# Calculating total votes in the election
totalVotes = 0

for vote in voteCountDF["Votes"]: # for each vote in voteCountDF
    totalVotes += vote # add it to totalVotes

# print(voteCount)
for row in voteCount: # for each row in voteCount
    row["Vote Share"] = (row["Votes"] * 100 )/totalVotes # add "Vote Share" key

# Converting voteCount to a data frame
voteCountDF = pd.DataFrame.from_dict(voteCount)

# Displaying voteCountDF
voteCountDF
```

Now, display the top 5 parties and find out the number of votes received by them

```python
# Getting unique parties
parties = set(df["Party"].tolist())

# Initializing partyVoteCount
partyVoteCount = {
    "Party": [],
    "Votes": []
}

for party in parties: # for each party
    # Calculating total votes
    totalVotes = 0
    # getting all candidate's vote from a party
    candVotes = df[df["Party"] == party]["Total Votes"]
    # adding all candidate's vote to totalVotes
    for candVote in candVotes:
        totalVotes += candVote
    # appending party name and vote count in partyVoteCount
    partyVoteCount["Party"].append(party)
    partyVoteCount["Votes"].append(totalVotes)
    
# Coverting partyVoteCount into a data frame
partyVoteCountDF = pd.DataFrame.from_dict(partyVoteCount)

# sorting partyVoteCountDF in descending order
sortedPartyVoteCountDF = partyVoteCountDF.sort_values(by=["Votes"], ascending=False)

# displaying top 5 voted parties
sortedPartyVoteCountDF.head(5)
```

From the candidates who lost, list 10 candidates who got the highest percent votes.

```python
lost = df[df['Result']=='Lost']
lost[['% of Votes']] = lost[['% of Votes']].apply(pd.to_numeric)
lost.sort_values(by="% of Votes", inplace=True, ascending=False, ignore_index=True)
lost.head(10)
```

### These techniques help to understand the data that we have but we like something a bit more visual right? That's where matplotlib comes to our aid.

Make a piechart to show the votes recieved by the top 5 candidates in the elections.

```python
topFive = df.sort_values(by="Total Votes", ascending=False).head(5)
topFiveList = topFive["Total Votes"].tolist()
labelsList = topFive["Candidate"].tolist()
plt.pie(topFiveList, labels=labelsList)
plt.show
```

Make a bar graph to see the postal votes received by the parties in Karnataka

```python
karnatakaDF = df[df["State"] == "Karnataka"]
karnatakaParties = list(set(karnatakaDF["Party"].tolist()))
postalVotes = []
for party in karnatakaParties:
    partyPostalVotes = 0
    candVotes = karnatakaDF[karnatakaDF["Party"] == party]["Postal Votes"]
    for candVote in candVotes:
        if (candVote != "-"):
            partyPostalVotes += float(candVote)
    postalVotes.append(partyPostalVotes)
fig = plt.figure(figsize = (20, 5))
plt.bar(karnatakaParties, postalVotes)
plt.xticks(rotation='vertical')
plt.show()
```
